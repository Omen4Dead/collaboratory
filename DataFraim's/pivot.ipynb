## Пример
Считаем кол-во `aggfunc='count'` фильмов `values='movieId'` с разбивкой по рейтингу `columns='rating'` и пользователям `index='userId'`.
Заменяем для наглядности `NaN` на `0`
Выводим строку и столбец итогов `margins=True`

```python
df.pivot_table(values='movieId', 
               index='userId', 
               columns='rating', 
               aggfunc='count', 
               fill_value=0,
               margins=True)
```
