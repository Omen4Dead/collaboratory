```python
import pandas as pd
```

```python
city_df = pd.read_csv('city.csv', sep=';')
city_df
```

```python
city_df.groupby('CountryCode', as_index=False)[['Population']].max()
```

```python
city_df.groupby('CountryCode', 
                as_index=False).filter(lambda x: 
                                       len(x['Population'])> 100).groupby('CountryCode').max()
```

```python
city_df.groupby('CountryCode').apply(lambda x: x['Population'].max() if len(x['CountryCode']) > 100 else 0)
```
