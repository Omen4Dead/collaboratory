```python
import pandas as pd
```

```python
cities = pd.Series(
    {'Dnepr': 1000000, 'Kiev': 3000000, 'Paris': 2300000, 'Berlin': 3800000},
    index=['Dnepr', 'Kiev', 'Paris', 'Berlin', 'Milan']
)
cities
```

```python
cities.isnull()
```

```python
cities.notnull()
```

```python
cities > 1000000
```

```python
cities['Dnepr']
```

```python
cities[cities.notnull()]
```

`&` - and

`|` - or

`~` - not

```python
cities[ (cities > 1000000) & (cities < 3000000) ]
```

```python
cities[~(cities > 3000000)]
```
