```python
df = pd.read_csv('city.csv', sep=';')
df.head()
```

```python
data = df.groupby('CountryCode')[['Population']].sum().sort_values(by='Population', 
                                                                   ascending=False).head(10)
data
```

```python
plt.bar(data.index, data.Population)
```

```python
plt.barh(data.index, data.Population)
```

```python
plt.barh(data.index, data.Population)
plt.legend(['Population'])
```

```python
plt.bar(data.index, data.Population)
plt.legend(['Population'])
plt.xlabel('Country_x')
plt.ylabel('Population_y')
plt.title('Top 10 country population')
plt.grid()
```

```python
data.plot(kind='bar', grid=True)
```

```python
data.plot(kind='pie', y='Population')
```

```python
plt.pie(data.Population, labels=data.index, autopct='%1.1f%%')
plt.show()
```
