```python
import pandas as pd
```

```python
df = pd.read_csv('result.csv')
df
```

```python
df.groupby('month').sum().sort_values(by=['total'], ascending=False)[['total']]
```

```python
df.order_date = pd.to_datetime(df.order_date, format=r'%Y-%m-%d %H:%M:%S')
df
```

```python
df.info(memory_usage='deep')
```

```python
df['hour'] = df.order_date.dt.hour
```

```python
df['day_name'] = df.order_date.dt.day_name()
```

```python
df
```

```python
res_hours = df.groupby('hour').agg(['sum', 'count'])['total']
res_hours
```

```python
res_hours.sort_values('sum', ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 8))
plt.bar(res_hours.index, res_hours['sum'])
plt.grid()
plt.xticks(range(0, 24, 1))
plt.yticks(range(0, int(round(res_hours.max()[0])), 200000))
plt.show()
```
