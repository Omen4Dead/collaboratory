```python
import pandas as pd
```

```python
df = pd.read_csv('result.csv', sep=',')
df.head()
```

```python
df.info(memory_usage='deep')
```

```python
df.describe()
```

```python
df.rename(columns={
    'Order ID': 'order_id',
    'Product': 'product',
    'Quantity Ordered': 'qt',
    'Price Each': 'price',
    'Order Date': 'order_date',
    'Purchase Address': 'address'
}, inplace=True)
df
```

```python
df.drop( df[df.qt.str.contains('Quantity Ordered')].index, inplace=True )
df
```

```python
df.qt = df.qt.astype('int')
df.price = df.price.astype('float')
```

```python
df.info(memory_usage='deep')
```

```python
# 04/19/19 08:46
df.order_date = df.order_date.apply(lambda x: pd.to_datetime(arg=x, format=r'%m/%d/%y %H:%M'))
```

```python
df.info(memory_usage='deep')
```

```python
df['total'] = df.price * df.qt
```

```python
df.columns
```

```python
df2 = df[['order_id', 'product', 'qt', 'price', 'total', 'order_date', 'address']].copy(deep=True)
```

```python
df2.to_csv('result.csv', sep=',', index=False)
```
